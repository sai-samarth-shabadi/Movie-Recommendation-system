```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#load the dataset
movies_data = pd.read_csv('movies.csv')
```

```python
#first 5 rows of data
movies_data.head()
```

```python
#size of dataset
movies_data.shape
```

```python
#selecting the relevent features for recommendation
selected_features = ['genres', 'keywords', 'title', 'cast', 'director']
```

```python
#replacing the null values with null string
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
#Combining all the selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
#Converting the text data into feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#getting the similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#Getting the movie name from the user
movie_name = input("Enter your favourite movie name: ")

#creating the list of movie names
list_of_all_titles = movies_data['title'].tolist()

#finding the close match for the movie given by user
close_matches_list = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = close_matches_list[0]

#finding the index of the movie using title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

#getting the list of similar movies
simiilarity_score = list(enumerate(similarity[index_of_the_movie]))

#Sorting the movies based on similarity score
sorted_similar_movies = sorted(simiilarity_score, key = lambda x:x[1], reverse=True)

#printing the name of similar movies based on the index
print("Movies recommended for you: \n")

i=1
for movie in sorted_similar_movies:
    similar_movie = movies_data[movies_data.index == movie[0]]['title'].values[0]
    if(i<=20):
        print(i,'.',similar_movie)
        i+=1
```
